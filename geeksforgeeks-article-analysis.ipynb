```python
import os
import numpy as np
import pandas as pd


df = pd.read_csv('/kaggle/input/geeksforgeeks-articles/articles.csv')
df.dropna(inplace = True)
```

## 1) How many authors do we have in the dataset?

```python
data = df.values

authors = set()
for i in range(len(data)):
    authors.add(data[i][1])

len(authors)
```

## 2) How many easy articles are publised on GeeksforGeeks?

```python
easy_art = []
easy_art = df[df['category'] == 'easy']
len(easy_art)
```

## 3) Which category is having most number of articles?

```python
max(df['category'])
```

## 4) What percent of articles are in expert category out of total number of articles?

```python
total = df['category']
expert = df[df['category'] == 'expert']
round((len(expert)/len(total))*100)
```

## 5) Who has written most number of articles after GeeksforGeeks?

## 6) Who has written most number of articles in Expert Category after GeeksforGeeks?

## 7) Which day of the month has most number of articles published?

```python
day = []
for i in range(len(data)):
    if(len(data[i][2].split(' ')) == 3):
        day.append(data[i][2].split(' ')[0])
    else:
        day.append(0)
```

```python
df['date'] = day
df
```

```python
df.groupby('date').size().sort_values(ascending = False).head().plot(kind = 'bar')
```

## 8) Which day of July has most number of articles published?

```python
month = []
for i in range(len(data)):
    if(len(data[i][2].split(' ')) == 3):
        month.append(data[i][2].split(' ')[1][:-1])
    else:
        month.append(0)
```

```python
df['month'] = month
df
```

```python
df[df['month'] == 'Jul'].groupby('date').size().sort_values(ascending = False).plot(kind = 'line')
```

## 9) Which month has least number of articles published?

```python
df.groupby('month').size().sort_values().head().plot(kind = 'pie')
```

## 10) Growth Rate of articles written 2021 as compare to 2020?

```python
year = []
for i in range(len(data)):
    if(len(data[i][2].split(' ')) == 3):
        year.append(data[i][2].split(' ')[2])
    else:
        year.append(0)
```

```python
df['year'] = year
df
```

```python
y20 = len(df[df['year'] == '2020'])
y21 = len(df[df['year'] == '2021'])

((y21-y20)/y20)*100
```
